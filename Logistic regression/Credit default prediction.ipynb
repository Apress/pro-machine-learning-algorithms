# Predicting credit default

```python
import pandas as pd
```

```python
data=pd.read_csv('F:/course/Logistic regression/credit_training.csv')
```

```python
data.head()
```

```python
# seems like the first column is not needed
data=data.drop(['Unnamed: 0'],axis=1)
```

```python
data.head()
```

```python
data.describe()
```

```python
# Seems like monthlyincome & numberofdependents have some missing values
# Also, there seem to be outliers across variables
# Lets fix the missing values first & then outliers
```

```python
# There are multiple ways in which we can fix missing values
# for now, lets just impute missing values with median of that column
```

```python
data['MonthlyIncome']=data['MonthlyIncome'].fillna(value=data['MonthlyIncome'].median())
```

```python
data['MonthlyIncome'].describe()
```

```python
data['NumberOfDependents']=data['NumberOfDependents'].fillna(value=data['NumberOfDependents'].median())
```

```python
data.describe()
```

```python
vars=["RevolvingUtilizationOfUnsecuredLines","NumberOfTime30to59DaysPastDueNotWorse","DebtRatio","MonthlyIncome","NumberOfTimes90DaysLate","NumberRealEstateLoansOrLines","NumberOfTime60to89DaysPastDueNotWorse"]
```

```python
t="RevolvingUtilizationOfUnsecuredLines"
data[t].quantile(0.95)
```

```python
import numpy as np
x=data[t].quantile(0.95)
data[t+"outlier_flag"]=np.where(data[t]>x,1,0)
```

```python
data.describe()
```

```python
for t in vars:
    x=data[t].quantile(0.95)
    data[t+"outlier_flag"]=np.where(data[t]>x,1,0)
    data[t]=np.where(data[t]>x,x,data[t])
    
```

```python
data.describe()
```

```python
data['DebtRatio'].quantile(np.arange(0,1,0.05))
```

```python
data['DebtRatio_newoutlier']=np.where(data['DebtRatio']>1,1,0)
data['DebtRatio']=np.where(data['DebtRatio']>1,1,data['DebtRatio'])
```

```python
data.describe()
```

```python
import numpy as np
sample=np.random.randint(0,data.shape[0],np.around(data.shape[0]*0.8))
```

```python
train=data.loc[sample]
```

```python
test=data.drop(data.index[sample])
```

```python
# importing the package for logistic regression
import statsmodels.formula.api as smf
```

```python
data.columns
```

```python
est = smf.logit(formula='SeriousDlqin2yrs~RevolvingUtilizationOfUnsecuredLinesoutlier_flag+NumberOfTime30to59DaysPastDueNotWorseoutlier_flag+DebtRatiooutlier_flag+MonthlyIncomeoutlier_flag+NumberOfTimes90DaysLateoutlier_flag+NumberRealEstateLoansOrLinesoutlier_flag+RevolvingUtilizationOfUnsecuredLines+age+NumberOfTime30to59DaysPastDueNotWorse+DebtRatio+MonthlyIncome+MonthlyIncome+       NumberOfOpenCreditLinesAndLoans+ NumberOfTimes90DaysLate+       NumberRealEstateLoansOrLines+ NumberOfTime60to89DaysPastDueNotWorse+       NumberOfDependents+DebtRatio_newoutlier+NumberOfTime60to89DaysPastDueNotWorseoutlier_flag',data=train)
```

```python
est2=est.fit()
```

```python
print(est2.summary())
```

```python
pred=est2.predict(test)
```

```python
pred
```

```python
test['prediction']=pred
```

```python
test=test.round({'prediction':2})
test.head()
```

```python
test['prediction_rank']=pd.qcut(test['prediction'].rank(method='first').values, 10).codes + 1 
test['prediction_rank'].describe()
print(test.groupby("prediction_rank").agg({"prediction": [np.mean, np.sum,len]}).rename(columns={'mean':'mean_predicted_default','sum':'total_defaults','len':'total_observations'}).to_string())
```

```python
test.head()
```

```python
test['SeriousDlqin2yrs'].mean()
```

```python
from sklearn.metrics import roc_auc_score
y_true = np.array(test['SeriousDlqin2yrs'])
y_scores = np.array(test['prediction'])
roc_auc_score(y_true, y_scores)
```
